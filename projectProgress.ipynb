```python
import matplotlib.pyplot as plt
all_tasks_fr = [
    "Construction des données", "Fonction de connexion", "Fonction d'inscription",
    "Fonction de réinitialisation du mot de passe", "FAQ (Partie complétée)",
    "FAQ (Ajout de nouvelles questions)", "Playlist (Frontend complété)",
    "Développement backend (Playlist)", "Création de playlist", "Ajout de réunions",
    "Résolution des problèmes avec le professeur et $_SESSION",
    "Page de localisation", "Page Qui Sommes Nous", "Page Capteur", "Page Information",
    "Gestion du niveau d'accès", "Amélioration de la convivialité", "Optimisation artistique"
]

all_weights = [3, 4, 4, 4, 3, 1, 4, 2, 1, 4, 0.25, 0.25, 0.5, 2, 0.5, 2, 1, 1]


all_percentages = [
    100, 95, 95, 90, 90, 0, 90, 0, 0, 90, 0, 0, 0, 0, 0, 0, 0, 0
]


weighted_total_progress_updated = sum(p * w for p, w in zip(all_percentages, all_weights)) / sum(all_weights)


fig, ax = plt.subplots(figsize=(14, 10))
ax.barh(all_tasks_fr, all_percentages, color='teal')
ax.set_xlabel('Pourcentage de complétion (%)')
ax.set_title('Graphique de progression du projet global après ajustements')
ax.set_xlim(0, 100)


ax.axvline(x=weighted_total_progress_updated, color='red', linestyle='--', label=f'Progression globale moyenne: {weighted_total_progress_updated:.2f}%')
ax.legend()


for i, (task, pct) in enumerate(zip(all_tasks_fr, all_percentages)):
    ax.text(pct + 1, i, f"{pct}% (Poids: {all_weights[i]})", color='darkblue', va='center')


plt.show()


weighted_total_progress_updated
```
